# Divide function

```python
def divide(number1, number2):
  try:
    number1_integer = int(number1)
    number2_integer = int(number2)
    return number1_integer / number2_integer
  except ValueError:
    return "Only integers are allowed. Try again"
  except ZeroDivisionError:
    return "Can not divide numbers by zero."
```

# Number input from a user

```python
user_number1 = input("Please, enter number1: ")
user_number2 = input("Please, enter number2: ")
```

```python
print(divide(user_number1, user_number2))
```
