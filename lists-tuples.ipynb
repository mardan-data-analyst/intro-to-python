# Lists

```python
shopping_list1 = ["Apple", "Juice", "Ball", "Book"]
```

```python
shopping_list1.append("Water")
print(shopping_list1)
```

```python
shopping_list1.insert(2, "Banana")
print(shopping_list1)
```

```python
shopping_list1.remove("Apple")
print(shopping_list1)
```

```python
shopping_list1[3] = "Strawberry"
print(shopping_list1)
```

## Creating new list from existing one - my solution

```python
number_list = [1, 2, 3, 4, 5]
square_list1 = []
square_list2 = []
```

```python
for i in number_list:
  i *= i
  square_list1.append(i)
print(square_list1)
```

## Creating new list from existing one - program solution

```python
for num in number_list:
  square_list2.append(num**2)
```

```python
print(square_list2)
```

## Creating new list from existing one - list comprehension

```python
square_list3 = [num**2 for num in number_list]
```

```python
print(square_list3)
```

# Tuples

### Packing elements into tuple (creating tuple)

```python
shopping_list2 = ("Apple", "Juice", "Ball", "Book")
```

```python
print(shopping_list2)
```

### Unpacking tuple

```python
item1, item2, item3, item4 = shopping_list2
```

```python
print(item1)
print(item2)
print(item3)
print(item4)
```

## Getting elements with their indexes

```python
print(shopping_list2[0])
```

### Tuple slicing

```python
print(shopping_list2[1:3])
```
