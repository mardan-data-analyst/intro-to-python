# Positional arguments

```python
def argument_printer(arg_position_1, arg_position_2):
  print("Argument at position 1", arg_position_1)
  print("Argument at position 2", arg_position_2)
```

```python
argument_printer("value1", "value2")
```

```python
def greeting(first_name, last_name, auto_correction):
  if auto_correction == True:
    capitalized_first_name = first_name.capitalize()
    capitalized_last_name = last_name.capitalize()
    print("Hello,", capitalized_first_name, capitalized_last_name)
  else:
    print("Hello,", first_name, last_name)
```

```python
greeting("deniZ", "kaYa", auto_correction=True)
```

# Keyword arguments

```python
def car(brand, model, colour):
  print("My favourite car is the", brand, model, "in the colour of", colour)
```

```python
car("Audi", "R8", "black")
```

```python
car(colour = "black", model = "R8",  brand="Audi")
```

# Scopes

```python
number = 13 # global variable

def change_number():
  number = 4 # local variable
```

```python
change_number()
print(number)
```

```python
def change_number():
  global number
  number = 4
```

```python
change_number()
print(number)
```
