# Read mode

If the file is in the same directory with Python file, file name with file extension and mode are fiven as arguments to the open() function. If not specified, default mode is read.

```python
message_file = open("message.txt", "r")
```

```python
content = message_file.read()
message_file.close()
```

```python
print(content)
```

# Write mode

```python
hello_file = open("hello.txt", "w")
```

```python
hello_file.write("Hello, world!")
hello_file.close()
```

# Saving program output to the file

```python
import random
```

```python
numbers_file1 = open("number_history1.txt", "w")

while True:
  random_number = random.randrange(1000)
  print(random_number)
  numbers_file1.write(str(random_number))
  numbers_file1.write("\n")

  if random_number == 777:
    print("Found!")
    numbers_file1.write("Found!")
    numbers_file1.close()
    break
```

# Opening files using with command

```python
with open("number_history2.txt", "w") as numbers_file2:
  while True:
    random_number = random.randrange(1000)
    print(random_number)
    numbers_file2.write(str(random_number))
    numbers_file2.write("\n")
    
    if random_number == 777:
      print("Found!")
      numbers_file2.write("Found!")
      break
```
