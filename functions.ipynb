```python
def add_two(number_1, number_2):
  print(number_1 + number_2)
```

```python
add_two(5, 5)
```

```python
result = add_two(3, 3)
print(result)
```

## return keyword

```python
def add_two_num(number_1, number_2):
  answer = number_1 + number_2
  return answer
```

```python
result = add_two_num(8, 6)
print(result)
```
